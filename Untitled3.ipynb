```python
from google.colab import files
import cv2
from matplotlib import pyplot as plt

# Upload gambar manual
uploaded = files.upload()

# Ambil nama file yang diupload (asumsi hanya 1 file)
img_path = list(uploaded.keys())[0]

# Load gambar sebagai grayscale
gray_img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# Tampilkan gambar grayscale
plt.imshow(gray_img, cmap='gray')
plt.axis('off')
plt.title('Grayscale Image')
plt.show()
```

```python
gray_rgb = cv2.cvtColor(gray_img, cv2.COLOR_GRAY2RGB)
```

```python
# Load model DDColor
model_dir = snapshot_download('damo/cv_ddcolor_image-colorization', cache_dir='./modelscope')
colorizer = pipeline('image-colorization', model=model_dir)

# Convert grayscale ke 3 channel RGB karena pipeline DDColor membutuhkan 3 channel
gray_rgb = cv2.cvtColor(gray_img, cv2.COLOR_GRAY2RGB)

#  Colorize gambar
result = colorizer(gray_rgb)
colorized_img = result[OutputKeys.OUTPUT_IMG]
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(gray_img, cmap='gray')
plt.axis('off')
plt.title('Original Grayscale')

plt.subplot(1,2,2)
plt.imshow(colorized_img)
plt.axis('off')
plt.title('Colorized Image')
plt.show()
```

```python
import numpy as np

# Convert ke HSV untuk adjust saturasi
hsv_img = cv2.cvtColor((colorized_img*255).astype(np.uint8), cv2.COLOR_RGB2HSV)
h, s, v = cv2.split(hsv_img)

# Naikkan saturasi 1.2x supaya warna lebih kuat tapi natural
s = np.clip(s * 1.2, 0, 255).astype(np.uint8)

hsv_img = cv2.merge([h, s, v])
colorized_img_adjusted = cv2.cvtColor(hsv_img, cv2.COLOR_HSV2RGB)


# Tampilkan hasil
plt.imshow(colorized_img_adjusted)
plt.axis('off')
plt.title('Colorized with Adjusted Saturation')
plt.show()
```
